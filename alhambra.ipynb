```python

def descomprimir_archivo_gz(archivo_gz, archivo_destino):
    with gzip.open(archivo_gz, 'rb') as archivo_comprimido:
        with open(archivo_destino, 'wb') as archivo_dest:
            shutil.copyfileobj(archivo_comprimido, archivo_dest)

descomprimir_archivo_gz(ARCHIVO_GZ, ARCHIVO_DESTINO)
```
